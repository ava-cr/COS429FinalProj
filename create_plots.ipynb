```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

# Create some data for the bars
categories = np.array(['White', 'Black', 'East Asian', 'South Asian', 'Other'])
x_pos = np.arange(len(categories))
values1 = np.array([42.50885777, 19.0990383, 14.48456217, 16.77071031, 7.136831449])
values2 = np.array([36.43617021276596, 17.553191489361720, 20.079787234042552, 16.090425531914892, 9.840425531914894])
values3 = np.array([36.6091030789826, 18.60776439089692, 19.481927710843372, 15.796519410977242, 9.504685408299867])
bw = np.array([38.75675675675676, 18.783783783783782, 17.54054054054054, 15.864864864864865, 9.054054054054054])


# Set the width of the bars
bar_width = 0.2
# bar_width = 0.35

# Create a figure and an axis object
fig, ax = plt.subplots()

ax.bar(x_pos, values1, bar_width, label='Dataset', color = cm.Blues(255))
ax.bar(x_pos + bar_width, values2, bar_width, label='Baseline', color = cm.Blues(100))
ax.bar(x_pos + 2 * bar_width, values3, bar_width, label='All Filters', color = cm.Blues(200))
ax.bar(x_pos + 3 * bar_width, bw, bar_width, label='B&W Filters', color = cm.Blues(150))


# looking at all classes besides South Asian, color filters does slightly improve upon the Baseline
# in terms of misclassifying a proportionate amount of every ethnicity. This can be seen in the graph
# by the fact that the "Color Filters" bars get closer to the Dataset bars than the Baseline bars do.

ax.set_xticks(x_pos + bar_width*1.5)
ax.set_xticklabels(categories)

# Add labels, legend and a title
ax.set_ylabel('Percentage')
ax.set_title('Ethnicity Breakdown')
ax.legend()

# Display the plot
plt.show()
```
